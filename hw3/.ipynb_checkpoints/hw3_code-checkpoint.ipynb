```python
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import numpy as np
import pandas as pd
import datetime
```

## Q2 Isocontours of Normal Distributions

```python
def plot_q2_1(mean, cov, name):
    x, y = np.mgrid[-1+mean[0] : 1+mean[1] : .01, -1+mean[1]:1+mean[1]:.01]
    pos = np.dstack((x, y))
    rv = multivariate_normal(mean, cov)
    plt.title(name)
    plt.contourf(x, y, rv.pdf(pos))
    plt.show()
def plot_q2_2(mean, cov, name):
    x, y = np.mgrid[-3+(mean[0][0]+mean[1][0])/2 : 3+(mean[0][0]+mean[1][0])/2 : .01,
                    -3+(mean[0][1]+mean[1][1])/2 : 3+(mean[0][1]+mean[1][1])/2 : .01]
    pos = np.dstack((x, y))
    rv1 = multivariate_normal(mean[0], cov[0])
    rv2 = multivariate_normal(mean[1], cov[1])
    plt.title(name)
    plt.contourf(x, y, rv1.pdf(pos)-rv2.pdf(pos))
    plt.show()
```

```python
plot_q2_1([1,1], [[1,0],[0,2]], '2a')
plot_q2_1([-1,2],[[2,1],[1,3]], '2b')
plot_q2_2([[0,2], [2,0]], [[[2, 1], [1,1]], [[2, 1], [1,1]]], '2c')
plot_q2_2([[0,2], [2,0]], [[[2, 1], [1,1]], [[2, 1], [1,3]]], '2d')
plot_q2_2([[1,1], [-1,-1]], [[[2, 0], [0,1]], [[2, 1], [1,2]]], '2d')
```

## Q3 Eigenvectors of the Gaussian Covariance Matrix

```python
np.random.seed(121)
x1 = np.random.normal(3, 3, 100)
x2 = x1/2 + np.random.normal(4, 2, 100)
```

```python
# 3a
mean_3a = (np.mean(x1), np.mean(x2))
mean_3a
```

```python
# 3b
cov_3b = np.cov((x1, x2))
cov_3b
```

```python
# 3c
eig_3c = np.linalg.eig(cov_3b)
eig_3c
print('eig. value', eig_3c[0][0], ': eig. vector', eig_3c[1][0])
print('eig. value', eig_3c[0][1], ': eig. vector', eig_3c[1][1])
```

```python
# 3d
plt.axis((-15,15, -15,15))
plt.scatter(x1, x2);
dx=eig_3c[0][0]*eig_3c[1][0][0]/np.linalg.norm(eig_3c[1][0], ord=1)
dy=eig_3c[0][0]*eig_3c[1][0][1]/np.linalg.norm(eig_3c[1][0], ord=1)
plt.arrow(x=mean_3a[0], y=mean_3a[1], dx=dx, dy=dy, width = 0.5, color='y');
dx=eig_3c[0][1]*eig_3c[1][1][0]/np.linalg.norm(eig_3c[1][1], ord=1)
dy=eig_3c[0][1]*eig_3c[1][1][1]/np.linalg.norm(eig_3c[1][1], ord=1)
plt.arrow(x=mean_3a[0], y=mean_3a[1], dx=dx, dy=dy, width = 0.5, color='y');
```

```python
# 3e
UT = np.array([eig_3c[1][0], eig_3c[1][1]])
X = np.array([(x1-mean_3a[0]), (x2-mean_3a[1])])
xr = UT@X
plt.axis((-15,15, -15,15));
plt.scatter(xr[0], xr[1]);
```

## Q7 Gaussian Classifiers for Digits and Spam

```python
# load the data
mnist = loadmat('./mnist-data/mnist_data.mat')
mnist_train_x = mnist['training_data']
mnist_train_y = mnist['training_labels']
mnist_train_x = (mnist_train_x-np.mean(mnist_train_x))/np.std(mnist_train_x)

# 7a
idx_7a = []
mean_7a = []
cov_7a = []
for i in range(10):
    idx = np.where(mnist_train_y == i)[0]
    mean = np.mean(mnist_train_x[idx], axis=0)
    cov = np.cov(mnist_train_x[idx], rowvar=False, bias=True)
    idx_7a.append(idx)
    mean_7a.append(mean)
    cov_7a.append(cov)
```

```python
# 7b
plt.title('Cov matrix For Digit 9');
plt.imshow(cov_7a[9]);
```

```python
# Load data
def split(dir_, val_amt):
    data = loadmat(dir_)
    n = data["training_data"].shape[0]
    shuffle = np.random.permutation(n)
    data_x, data_y = data["training_data"][shuffle], data["training_labels"][shuffle]
    data_val_x, data_val_y = data_x[:val_amt,:], data_y[:val_amt,:]
    data_train_x, data_train_y = data_x[val_amt:,:], data_y[val_amt:,:]
    return data_train_x, data_train_y, data_val_x, data_val_y

mnist_train_x, mnist_train_y, mnist_val_x, mnist_val_y = split('./mnist-data/mnist_data.mat', 10000)
mnist_mean = np.mean(mnist_train_x)
mnist_std = np.std(mnist_train_x)
mnist_train_x = (mnist_train_x - mnist_mean)/mnist_std
mnist_val_x = (mnist_val_x - mnist_mean)/mnist_std
```

```python
# 7c
def prepare_LDA_mnist(train_amt):
    mean_7c = []
    cov_7c = []
    idx_7c = []
    for i in range(10): #0-9 digits
        idx = np.where(mnist_train_y[:train_amt,:] == i)[0]
        mean = np.mean(mnist_train_x[idx], axis=0)
        cov = np.cov(mnist_train_x[idx], rowvar=False, bias=True)
        mean_7c.append(mean)
        cov_7c.append(cov*len(idx))
        idx_7c.append(idx)
    cov_7c = sum(cov_7c)/train_amt #one covariance matrix for all classes
    prior_7c = np.array([len(idx)/train_amt for idx in idx_7c])
    return mean_7c, cov_7c, prior_7c

def generate_LDA_mnist(mean, cov):
    LDA_gaussians = []
    for mu in mean:
        LDA_gaussians.append(multivariate_normal(mu, cov, allow_singular=True))
    return LDA_gaussians

def predict_LDA_mnist(samples, gaussian, priors):
    pdfs = np.array([gauss.logpdf(samples) for gauss in gaussian]).T
    posteriors = pdfs + np.log(priors)
    predictions = posteriors.argmax(axis=1)
    return predictions

def train_LDA_mnist(amts):
    train_errs, val_errs, gaussians, priors = [],[],[],[]
    for amt in amts:
        means, covs, priors = prepare_LDA_mnist(amt)
        gaussians = generate_LDA_mnist(means, covs)
        train_predictions = predict_LDA_mnist(mnist_train_x, gaussians, priors).reshape(-1)
        
        val_predictions = predict_LDA_mnist(mnist_val_x, gaussians, priors).reshape(-1)
        train_errs.append(np.count_nonzero(train_predictions != mnist_train_y.reshape(-1))/len(mnist_train_y))
        val_errs.append(np.sum(val_predictions != mnist_val_y.reshape(-1))/len(mnist_val_y))
    plt.plot(amts,train_errs)
    plt.plot(amts,val_errs)
    plt.legend(["training error","validation error"])
    plt.title("LDA on MNIST")
    plt.xlabel("training size")
    plt.ylabel("error")
    return gaussians, priors

amts = [100, 200, 500, 1000, 2000, 5000, 10000, 30000, 50000]
gaussians, priors = train_LDA_mnist(amts)
```

```python
mnist_test_x = loadmat("mnist-data/mnist_data.mat")["test_data"]
def results_to_csv(y_test, setname):
    y_test = y_test.astype(int)
    df = pd.DataFrame({'Category': y_test})
    df.index += 1  # Ensures that the index starts at 1. 
    df.to_csv(setname+'-'+str(datetime.datetime.now().time())+ \
              '-'+'submission.csv', index_label='Id')
def test_predictions_mnist():
    gaussians, priors = train_LDA_mnist([60000])
    test_predictions = generate_predictions_LDA(mnist_test_x,gaussians,priors)
    results_to_csv(test_predictions,"mnist")
test_predictions_mnist()
```
